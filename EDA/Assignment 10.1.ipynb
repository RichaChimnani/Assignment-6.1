```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

names = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")
```

```python
print(names.shape)
```

```python
names.head()
```

#1. Delete unnamed columns

```python
del names['Unnamed: 0']
```

```python
print("\n 1.Delete unnamed Columns\n",'-'*35, sep='')

names.head()
```

#2. Show the distribution of male and female

```python
gen_distribution =  names.pivot_table(index='Gender',
                         values='Count',
                         aggfunc=np.sum)

gen_distribution
```

```python
print("\n 2. Show the Distribution of male and female\n",'-'*45, sep='')


labels = 'Female', 'Male'
explode = (0.1, 0)  # explode the 1st slice


plt.pie(gen_distribution, explode=explode, labels=labels, autopct='%1.02f%%', startangle=90)


plt.axis('equal') # making the pie chart flat
plt.title("Male/Female Distribution")
```

 #3. Show the top 5 most preferred names

```python
print("\n 3. Show the top 5 most preferred names.\n",'-'*45, sep='')

top5_names = names.groupby(['Name'])['Count'].sum().nlargest(5)  
top5_names.head()
```

#4  What is the median name occurence in the dataset

```python

print("\n 4. What is the median name occurence in the dataset.\n",'-'*60, sep='')


names[names.Id == names.median().Id]['Name']
```

#5. Distribution of male and female born count by states

```python
print("\n 5. Distribution of male and female born count by states\n",'-'*65, sep='')

genders = names.pivot_table(index='State',
                         columns='Gender',
                         values='Count',
                         aggfunc=np.sum)
genders
```
